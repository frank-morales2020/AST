```python
# =============================================================================
# MICROGPT - The Complete Algorithm in Pure Python
# Author: Andrej Karpathy (with Colab-friendly modifications)
#
# "The most atomic way to train and run inference for a GPT in pure Python.
#  This file is the complete algorithm. Everything else is just efficiency."
# =============================================================================

import os
import math
import random
import time
from IPython.display import clear_output  # For Colab nice display

# ===== ADDED: Import ast_lefm (YOUR LIBRARY) =====
try:
    from ast_lefm.sieve import primes_up_to
    from ast_lefm.lefm import coherence, spectral_trap
    from ast_lefm.constants import UNIVERSAL_COHERENCE
    print("✓ ast_lefm loaded")
except ImportError:
    print("⚠️ ast_lefm not available")

# ===== ADDED: Simple Nested Learning Observer (watches only, does nothing) =====
class NestedObserver:
    def __init__(self):
        self.count = 0
    def watch(self, token_id):
        self.count += 1
    def report(self):
        print(f"\n[Nested Learning] Observed {self.count} tokens")

nested = NestedObserver()
# ===== END OF ADDITIONS =====

# Set random seed for reproducibility
#random.seed(42)
random.seed(123)
print("=" * 60)
print("🚀 MICROGPT - Training a Transformer from Scratch")
print("=" * 60)

# =============================================================================
# STEP 1: DATA PREPARATION - Download and load names dataset
# =============================================================================
print("\n📥 Step 1: Loading dataset...")

if not os.path.exists('input.txt'):
    import urllib.request
    names_url = 'https://raw.githubusercontent.com/karpathy/makemore/988aa59/names.txt'
    urllib.request.urlretrieve(names_url, 'input.txt')
    print("   Downloaded names.txt")

docs = [line.strip() for line in open('input.txt') if line.strip()]
random.shuffle(docs)
print(f"   Loaded {len(docs):,} names")
print(f"   Example names: {docs[:5]}")

# =============================================================================
# STEP 2: TOKENIZATION - Create character vocabulary
# =============================================================================
print("\n🔤 Step 2: Creating tokenizer...")

uchars = sorted(set(''.join(docs)))
BOS = len(uchars)  # Beginning of Sequence token
vocab_size = len(uchars) + 1
print(f"   Unique characters: {''.join(uchars)}")
print(f"   Vocabulary size: {vocab_size} (including BOS token)")

# =============================================================================
# STEP 3: AUTOGRAD ENGINE - Build automatic differentiation from scratch
# =============================================================================
print("\n🧮 Step 3: Building autograd engine...")

class Value:
    __slots__ = ('data', 'grad', '_children', '_local_grads')

    def __init__(self, data, children=(), local_grads=()):
        self.data = data
        self.grad = 0
        self._children = children
        self._local_grads = local_grads

    def __add__(self, other):
        other = other if isinstance(other, Value) else Value(other)
        return Value(self.data + other.data, (self, other), (1, 1))

    def __mul__(self, other):
        other = other if isinstance(other, Value) else Value(other)
        return Value(self.data * other.data, (self, other), (other.data, self.data))

    def __pow__(self, other):
        return Value(self.data**other, (self,), (other * self.data**(other-1),))

    def log(self): return Value(math.log(self.data), (self,), (1/self.data,))
    def exp(self): return Value(math.exp(self.data), (self,), (math.exp(self.data),))
    def relu(self): return Value(max(0, self.data), (self,), (float(self.data > 0),))

    def __neg__(self): return self * -1
    def __radd__(self, other): return self + other
    def __sub__(self, other): return self + (-other)
    def __rsub__(self, other): return other + (-self)
    def __rmul__(self, other): return self * other
    def __truediv__(self, other): return self * other**-1
    def __rtruediv__(self, other): return other * self**-1

    def backward(self):
        topo = []
        visited = set()

        def build_topo(v):
            if v not in visited:
                visited.add(v)
                for child in v._children:
                    build_topo(child)
                topo.append(v)

        build_topo(self)
        self.grad = 1
        for v in reversed(topo):
            for child, local_grad in zip(v._children, v._local_grads):
                child.grad += local_grad * v.grad

print("   ✓ Autograd engine ready!")

# =============================================================================
# STEP 4: MODEL ARCHITECTURE - Initialize transformer parameters
# =============================================================================
print("\n🏗️  Step 4: Building transformer architecture...")

# Hyperparameters
n_layer = 1      # Number of transformer layers
n_embd = 16      # Embedding dimension
block_size = 16  # Maximum context length
n_head = 4       # Number of attention heads
head_dim = n_embd // n_head

print(f"   Configuration:")
print(f"     - Layers: {n_layer}")
print(f"     - Embedding size: {n_embd}")
print(f"     - Context window: {block_size}")
print(f"     - Attention heads: {n_head}")

# Initialize weights with small random values
matrix = lambda nout, nin, std=0.08: [[Value(random.gauss(0, std)) for _ in range(nin)]
                                       for _ in range(nout)]

state_dict = {
    'wte': matrix(vocab_size, n_embd),      # Token embeddings
    'wpe': matrix(block_size, n_embd),       # Position embeddings
    'lm_head': matrix(vocab_size, n_embd),   # Output projection
}

# Add transformer layers
for i in range(n_layer):
    state_dict[f'layer{i}.attn_wq'] = matrix(n_embd, n_embd)
    state_dict[f'layer{i}.attn_wk'] = matrix(n_embd, n_embd)
    state_dict[f'layer{i}.attn_wv'] = matrix(n_embd, n_embd)
    state_dict[f'layer{i}.attn_wo'] = matrix(n_embd, n_embd)
    state_dict[f'layer{i}.mlp_fc1'] = matrix(4 * n_embd, n_embd)
    state_dict[f'layer{i}.mlp_fc2'] = matrix(n_embd, 4 * n_embd)

# Flatten all parameters
params = [p for mat in state_dict.values() for row in mat for p in row]
print(f"   Total trainable parameters: {len(params):,}")

# =============================================================================
# STEP 5: FORWARD PASS - Define the GPT computation
# =============================================================================
print("\n🔄 Step 5: Defining forward pass...")

def linear(x, w):
    """Linear transformation: y = x @ w.T"""
    return [sum(wi * xi for wi, xi in zip(wo, x)) for wo in w]

def softmax(logits):
    """Convert logits to probabilities"""
    max_val = max(val.data for val in logits)
    exps = [(val - max_val).exp() for val in logits]
    total = sum(exps)
    return [e / total for e in exps]

def rmsnorm(x):
    """Root Mean Square Layer Normalization"""
    ms = sum(xi * xi for xi in x) / len(x)
    scale = (ms + 1e-5) ** -0.5
    return [xi * scale for xi in x]

def gpt(token_id, pos_id, keys, values):
    """GPT forward pass for a single token"""
    # Embeddings
    tok_emb = state_dict['wte'][token_id]
    pos_emb = state_dict['wpe'][pos_id]
    x = [t + p for t, p in zip(tok_emb, pos_emb)]
    x = rmsnorm(x)

    for li in range(n_layer):
        # ========== ATTENTION BLOCK ==========
        x_residual = x
        x = rmsnorm(x)

        # Compute Q, K, V
        q = linear(x, state_dict[f'layer{li}.attn_wq'])
        k = linear(x, state_dict[f'layer{li}.attn_wk'])
        v = linear(x, state_dict[f'layer{li}.attn_wv'])

        # Store in KV cache
        keys[li].append(k)
        values[li].append(v)

        # Multi-head attention
        x_attn = []
        for h in range(n_head):
            hs = h * head_dim
            q_h = q[hs:hs+head_dim]
            k_h = [ki[hs:hs+head_dim] for ki in keys[li]]
            v_h = [vi[hs:hs+head_dim] for vi in values[li]]

            # Scaled dot-product attention
            attn_logits = [sum(q_h[j] * k_h[t][j] for j in range(head_dim)) / head_dim**0.5
                          for t in range(len(k_h))]
            attn_weights = softmax(attn_logits)

            # Apply attention to values
            head_out = [sum(attn_weights[t] * v_h[t][j] for t in range(len(v_h)))
                       for j in range(head_dim)]
            x_attn.extend(head_out)

        # Project and residual
        x = linear(x_attn, state_dict[f'layer{li}.attn_wo'])
        x = [a + b for a, b in zip(x, x_residual)]

        # ========== MLP BLOCK ==========
        x_residual = x
        x = rmsnorm(x)
        x = linear(x, state_dict[f'layer{li}.mlp_fc1'])
        x = [xi.relu() for xi in x]
        x = linear(x, state_dict[f'layer{li}.mlp_fc2'])
        x = [a + b for a, b in zip(x, x_residual)]

    # Output projection
    logits = linear(x, state_dict['lm_head'])
    return logits

print("   ✓ Forward pass defined!")

# =============================================================================
# STEP 6: OPTIMIZER - Initialize Adam buffers
# =============================================================================
print("\n⚙️  Step 6: Setting up Adam optimizer...")

learning_rate, beta1, beta2, eps_adam = 0.01, 0.85, 0.99, 1e-8
m = [0.0] * len(params)  # First moment
v = [0.0] * len(params)  # Second moment

print(f"   Learning rate: {learning_rate}")
print(f"   Beta1: {beta1}, Beta2: {beta2}")

# =============================================================================
# STEP 7: TRAINING LOOP - Learn from data
# =============================================================================
print("\n🎯 Step 7: Training (this will take about 30 seconds)...")
print("-" * 60)

num_steps = 1000
start_time = time.time()

for step in range(num_steps):
    # Get a training example
    doc = docs[step % len(docs)]
    tokens = [BOS] + [uchars.index(ch) for ch in doc] + [BOS]
    n = min(block_size, len(tokens) - 1)

    # ===== ADDED: Watch tokens with nested observer =====
    for token_id in tokens:
        nested.watch(token_id)
    # ===== END OF ADDITION =====

    # Forward pass through sequence
    keys, values = [[] for _ in range(n_layer)], [[] for _ in range(n_layer)]
    losses = []

    for pos_id in range(n):
        token_id, target_id = tokens[pos_id], tokens[pos_id + 1]
        logits = gpt(token_id, pos_id, keys, values)
        probs = softmax(logits)
        loss_t = -probs[target_id].log()
        losses.append(loss_t)

    loss = (1 / n) * sum(losses)

    # Backward pass
    loss.backward()

    # Adam update
    lr_t = learning_rate * (1 - step / num_steps)
    for i, p in enumerate(params):
        m[i] = beta1 * m[i] + (1 - beta1) * p.grad
        v[i] = beta2 * v[i] + (1 - beta2) * p.grad ** 2
        m_hat = m[i] / (1 - beta1 ** (step + 1))
        v_hat = v[i] / (1 - beta2 ** (step + 1))
        p.data -= lr_t * m_hat / (v_hat ** 0.5 + eps_adam)
        p.grad = 0

    # Progress update
    if (step + 1) % 100 == 0 or step == 0:
        elapsed = time.time() - start_time
        print(f"   step {step+1:4d}/{num_steps} | loss: {loss.data:.4f} | time: {elapsed:.1f}s")

# =============================================================================
# STEP 8: INFERENCE - Generate new names!
# =============================================================================
print("\n" + "=" * 60)
print("✨ Step 8: Generating new names!")
print("=" * 60)

temperature = 0.5  # Lower = more predictable, higher = more creative
print(f"\nTemperature = {temperature} (controls randomness)")
print("-" * 40)

generated_names = []
for sample_idx in range(20):
    keys, values = [[] for _ in range(n_layer)], [[] for _ in range(n_layer)]
    token_id = BOS
    sample = []

    for pos_id in range(block_size):
        logits = gpt(token_id, pos_id, keys, values)
        probs = softmax([l / temperature for l in logits])
        token_id = random.choices(range(vocab_size), weights=[p.data for p in probs])[0]

        if token_id == BOS:
            break

        sample.append(uchars[token_id])

    name = ''.join(sample)
    generated_names.append(name)
    print(f"   {sample_idx+1:2d}. {name}")

print("\n" + "=" * 60)
print("🎉 Training complete! The model learned to generate names!")
print("=" * 60)
print("\n📊 Quick analysis:")
print(f"   - Total training time: {time.time() - start_time:.1f} seconds")
print(f"   - Final loss: {loss.data:.4f}")
print(f"   - Generated {len(generated_names)} new names")

# ===== ADDED: Report nested learning stats =====
nested.report()
# ===== END OF ADDITION =====

print("\n💡 Try changing:")
print("   - temperature (0.1 to 1.5) for more/less creativity")
print("   - n_layer to 2 for a deeper model")
print("   - num_steps to 2000 for more training")
print("\n🔬 The 'magic' of GPT is just math and repetition!")
```
