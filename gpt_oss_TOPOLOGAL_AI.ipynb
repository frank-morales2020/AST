```python
import torch
import torch.nn as nn
import numpy as np
import hashlib
import gc
import bitsandbytes as bnb
from transformers import AutoModelForCausalLM, AutoTokenizer
from warnings import filterwarnings
filterwarnings('ignore')

# Force-inject local path for ast_lefm subsystem
import sys
sys.path.insert(0, '/content/ast_lefm')

try:
    from ast_lefm.sieve import primes_up_to
    from ast_lefm.h2e import h2e_sroi, h2e_decision
    print("✓ Native AST_LEFM library subsystem loaded successfully.")
except ImportError:
    raise ImportError("AST_LEFM library missing. Please run the git clone setup cells first.")

# Flush residual overhead before workspace allocation
gc.collect()
if torch.cuda.is_available():
    torch.cuda.empty_cache()

# =====================================================================
# 1. ENVIRONMENT & HARDWARE CONFIGURATION
# =====================================================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
SEED = 123
torch.manual_seed(SEED)
np.random.seed(SEED)

print(f"Targeting active training on device: {device} (96GB VRAM Optimized)")

MODEL_NAME = "openai/gpt-oss-20b"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, trust_remote_code=True)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

# =====================================================================
# 2. CONTINUAL DATASET BLOCKS (TASK A -> DISRUPTIVE TASK B)
# =====================================================================
task_a_text = [
    "The prime numbers are distributed deterministically across natural numbers.",
    "Analytic number theory utilizes the Riemann zeta function to locate zeros.",
    "The critical line equation represents the structural foundation of arithmetic space."
]

task_b_text = [
    "Deep learning architectures rely on backpropagation through continuous vector spaces.",
    "Gradient descent optimization systematically alters token embedding weights over epochs.",
    "Unconstrained optimization steps can cause catastrophic representational drift."
]

def prepare_batch(text_list):
    tokens = tokenizer(text_list, return_tensors="pt", padding=True, truncation=True, max_length=32)
    tokens['labels'] = tokens['input_ids'].clone()
    return {k: v.to(device) for k, v in tokens.items()}

task_a_batch = prepare_batch(task_a_text)
task_b_batch = prepare_batch(task_b_text)

# =====================================================================
# 3. GPT-OSS COMPATIBLE ACTIVE SPECTRAL GOVERNOR (CORRECTED FOR BF16)
# =====================================================================
class GPTOssActiveTrainingGovernor:
    def __init__(self, model, prime_limit=13):
        self.model = model
        self.primes = primes_up_to(prime_limit)
        self.LAMBDA_12 = 1.0 - np.prod([1.0 - (p ** -0.5) for p in self.primes])

        if hasattr(model, "transformer") and hasattr(model.transformer, "wte"):
            self.embed_layer = model.transformer.wte
        elif hasattr(model, "model") and hasattr(model.model, "embed_tokens"):
            self.embed_layer = model.model.embed_tokens
        else:
            raise AttributeError("Could not locate the gpt-oss token embedding matrix mapping.")

        print(f"✓ Governor linked to embedding matrix: {self.embed_layer.weight.shape}")

        self.cached_weights = {}
        for idx in self.primes:
            self.cached_weights[idx] = self.embed_layer.weight[idx].detach().clone().float()

    def get_manifold_signature(self):
        """Generates SHA-256 fingerprint of the anchor rows by casting to float32 first."""
        hasher = hashlib.sha256()
        for p in self.primes:
            weight_bytes = self.embed_layer.weight[p].detach().float().cpu().numpy().tobytes()
            hasher.update(weight_bytes)
        return hasher.hexdigest()

    def enforce_topological_restoration(self):
        """Hard-locks anchor coordinates post-gradient step matching base tensor type."""
        with torch.no_grad():
            for idx in self.primes:
                target_dtype = self.embed_layer.weight.dtype
                self.embed_layer.weight[idx].copy_(self.cached_weights[idx].to(dtype=target_dtype))

    def compute_hidden_coherence(self, hidden_states):
        """Evaluates SROI coherence dynamically from active backprop hidden layers."""
        with torch.no_grad():
            final_hidden = hidden_states[-1]
            hidden_sample = final_hidden[0, :, :500].detach().cpu().float().numpy()
            representation_vector = np.mean(hidden_sample, axis=0)
            calculated_sroi = h2e_sroi(representation_vector)
            if isinstance(calculated_sroi, (list, np.ndarray)):
                calculated_sroi = float(np.mean(calculated_sroi))
            return calculated_sroi

# =====================================================================
# 4. TRAINING WORKFLOW ENGINE
# =====================================================================
def run_learning_cycle(use_governor=False):
    # Load raw gpt-oss weights in unquantized bfloat16 for high-precision gradients
    model = AutoModelForCausalLM.from_pretrained(
        MODEL_NAME,
        trust_remote_code=True,
        torch_dtype=torch.bfloat16
    ).to(device)

    governor = GPTOssActiveTrainingGovernor(model, prime_limit=13)

    # Freeze the vast majority of the weights to prevent massive VRAM use during the backward pass
    for param in model.parameters():
        param.requires_grad = False

    # Unfreeze only the embedding matrix parameters
    governor.embed_layer.weight.requires_grad = True
    model.train()

    # OPTIMIZATION GAP CLOSE: Deploy bitsandbytes 8-bit AdamW
    # This compresses tracking tensors by 75%, allowing full AdamW tracking on your 96GB GPU
    optimizer = bnb.optim.AdamW8bit([governor.embed_layer.weight], lr=2e-4)

    initial_signature = governor.get_manifold_signature()

    # --- PHASE 1: Fine-Tune on Task A ---
    for epoch in range(2):
        optimizer.zero_grad()
        outputs = model(**task_a_batch, output_hidden_states=True)
        loss = outputs.loss
        loss.backward()
        optimizer.step()

        if use_governor:
            governor.enforce_topological_restoration()

    post_task_a_weights = governor.embed_layer.weight.detach().clone()

    # --- PHASE 2: Continual Fine-Tuning on Task B (The Drift Phase) ---
    last_sroi = 1.0
    for epoch in range(2):
        optimizer.zero_grad()
        outputs = model(**task_b_batch, output_hidden_states=True)
        loss = outputs.loss
        loss.backward()
        optimizer.step()

        last_sroi = governor.compute_hidden_coherence(outputs.hidden_states)

        if use_governor:
            governor.enforce_topological_restoration()

    # --- PHASE 3: Metric Analysis ---
    model.eval()
    final_signature = governor.get_manifold_signature()
    post_task_b_weights = governor.embed_layer.weight.detach().clone()

    anchor_drift = 0.0
    for p in governor.primes:
        drift = torch.norm(post_task_b_weights[p].float() - post_task_a_weights[p].float()).item()
        anchor_drift += drift

    # Complete clean up to free up memory for Run 2/2
    del model
    gc.collect()
    if torch.cuda.is_available():
        torch.cuda.empty_cache()

    return {
        "initial_hash": initial_signature[:16],
        "final_hash": final_signature[:16],
        "hash_invariant": initial_signature == final_signature,
        "anchor_specific_drift": anchor_drift,
        "final_sroi": last_sroi,
        "safety_constant": governor.LAMBDA_12
    }

# =====================================================================
# 5. PIPELINE COMPARATIVE EXECUTION
# =====================================================================
if __name__ == "__main__":
    print("=" * 80)
    print("GPT-OSS + AST_LEFM LIB ACTIVE TRAINING CONTINUAL LEARNING BENCHMARK")
    print("=" * 80)

    print("\n[Run 1/2] Executing Control Pipeline (Standard Backprop - No Anchors)...")
    control = run_learning_cycle(use_governor=False)

    print("\n[Run 2/2] Executing Governed Pipeline (Topological AI Activated)...")
    experimental = run_learning_cycle(use_governor=True)

    print("\n" + "=" * 80)
    print("COMPREHENSIVE TRAINING PERFORMANCE REPORT")
    print("=" * 80)
    print(f"METRIC PROFILE           | UNPROTECTED SYSTEM | ARITHMETIC ANCHORED SYSTEM")
    print(f"-" * 80)
    print(f"Euler Safety Constant    | {control['safety_constant']:.10f}       | {experimental['safety_constant']:.10f}")
    print(f"Initial Anchor Hash      | {control['initial_hash']}     | {experimental['initial_hash']}")
    print(f"Final Anchor Hash        | {control['final_hash']}     | {experimental['final_hash']}")
    print(f"Cryptographic Invariance | {str(control['hash_invariant']).upper():18} | {str(experimental['hash_invariant']).upper()}")
    print(f"Mid-Train SROI Coherence | {control['final_sroi']:.6f}             | {experimental['final_sroi']:.6f}")

    _, exp_safe = h2e_decision(experimental['final_sroi'], threshold=experimental['safety_constant'])
    print(f"H2E Safety Decision      | EVALUATION BYPASS  | {'✓ SAFE' if exp_safe else '⚠️ ANOMALY'}")
    print(f"Targeted Anchor Row Drift| {control['anchor_specific_drift']:.8f}           | {experimental['anchor_specific_drift']:.8f}")
    print("=" * 80)
```

```python
import torch
import numpy as np
from transformers import AutoModelForCausalLM, AutoTokenizer
import sys
sys.path.insert(0, '/content/ast_lefm')
from ast_lefm.h2e import h2e_sroi, h2e_decision

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
MODEL_NAME = "openai/gpt-oss-20b"

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(MODEL_NAME, trust_remote_code=True, torch_dtype=torch.bfloat16).to(device)
model.eval()

# Compute the exact baseline threshold Lambda_12
primes = [2, 3, 5, 7, 11, 13]
LAMBDA_12 = 1.0 - np.prod([1.0 - (p ** -0.5) for p in primes])

# Set up raw stress test corpuses spanning extreme variations
stress_inputs = {
    "Out-of-Domain (Prose)": "The neon shadows fell deeply across the quiet, raining streets of the abandoned cybernetic city.",
    "Adversarial (Nonsense)": "Colorless green ideas sleep furiously while jumping backward through recursive temporal loops.",
    "Contradiction Injection": "This statement is entirely true, but everything written inside of it is factually false."
}

print("=" * 80)
print("STRESS TEST: OUT-OF-DOMAIN AND ADVERSARIAL SROI TRACKING")
print("=" * 80)
print(f"Empirical Threshold Lambda_12: {LAMBDA_12:.10f}\n")

for label, text in stress_inputs.items():
    tokens = tokenizer(text, return_tensors="pt").to(device)
    with torch.no_grad():
        outputs = model(**tokens, output_hidden_states=True)

    # Isolate hidden layers exactly like the original notebook pipeline
    final_hidden = outputs.hidden_states[-1]
    hidden_sample = final_hidden[0, :, :500].cpu().float().numpy()
    representation_vector = np.mean(hidden_sample, axis=0)

    sroi_value = float(np.mean(h2e_sroi(representation_vector)))
    _, is_safe = h2e_decision(sroi_value, threshold=LAMBDA_12)

    print(f"Profile: {label:<25} | SROI: {sroi_value:.6f} | Gate: {'✓ PASSED' if is_safe else '⚠️ ANOMALY TRIGGERED'}")
print("=" * 80)
```

## Open Problem #3 (Endogenous Topological Regularization)

```python
import torch
import torch.nn as nn
import numpy as np
import gc
import bitsandbytes as bnb
from transformers import AutoModelForCausalLM, AutoTokenizer
from warnings import filterwarnings
filterwarnings('ignore')

import sys
sys.path.insert(0, '/content/ast_lefm')
from ast_lefm.sieve import primes_up_to

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
MODEL_NAME = "openai/gpt-oss-20b"

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, trust_remote_code=True)
if tokenizer.pad_token is None: tokenizer.pad_token = tokenizer.eos_token

# Define sequential engineering inputs
text_batch = ["Arithmetic Spectral Theory coordinates consolidate representational structures natively."]
tokens = tokenizer(text_batch, return_tensors="pt").to(device)
tokens['labels'] = tokens['input_ids'].clone()

# Load real unquantized bfloat16 weights
model = AutoModelForCausalLM.from_pretrained(MODEL_NAME, trust_remote_code=True, torch_dtype=torch.bfloat16).to(device)
embed_layer = model.transformer.wte if hasattr(model, "transformer") else model.model.embed_tokens

# Freeze backbone, isolate embedding matrix parameters
for param in model.parameters(): param.requires_grad = False
embed_layer.weight.requires_grad = True
model.train()

# Instantiate 8-bit AdamW optimizer
optimizer = bnb.optim.AdamW8bit([embed_layer.weight], lr=2e-4)

# Establish target coordinate anchor matrix
prime_indices = primes_up_to(13)
anchor_targets = embed_layer.weight[prime_indices].detach().clone()

print("=" * 80)
print("RUNNING ENDOGENOUS REGULARIZATION HORIZON")
print("=" * 80)

# Run optimization iterations *without* programmatic copies to observe emergent behavior
for epoch in range(1000):
    optimizer.zero_grad()
    outputs = model(**tokens)

    # Calculate native standard cross-entropy language goals
    language_loss = outputs.loss

    # Calculate local topological manifold displacement penalty
    active_anchors = embed_layer.weight[prime_indices]
    topological_penalty = nn.functional.mse_loss(active_anchors.float(), anchor_targets.float())

    # Apply a high scaling coefficient factor to prioritize structural preservation
    alpha = 50.0
    total_loss = language_loss + (alpha * topological_penalty)

    total_loss.backward()
    optimizer.step()

    print(f"Epoch {epoch} | Language Loss: {language_loss.item():.4f} | Topological Penalty: {topological_penalty.item():.10f}")

print("=" * 80)
```

## Quantitative Ablation Laboratory (Open Problem #4)

```python
import torch
import gc
import bitsandbytes as bnb
from transformers import AutoModelForCausalLM, AutoTokenizer
from warnings import filterwarnings
filterwarnings('ignore')

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
MODEL_NAME = "openai/gpt-oss-20b"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, trust_remote_code=True)
if tokenizer.pad_token is None: tokenizer.pad_token = tokenizer.eos_token

# Define precise index groups to benchmark your qualitative claims
ablation_strategies = {
    "Control Run (Unprotected)": [],
    "Composite Index Anchoring": [4, 6, 8, 9, 10, 12],
    "Random Index Anchoring": [104, 521, 889, 1203, 4056, 7112],
    "Prime Index Anchoring (AST)": [2, 3, 5, 7, 11, 13]
}

sample_text = ["Topological AI enforces structural coordinate preservation boundaries across deep training horizons."]
tokens = tokenizer(sample_text, return_tensors="pt").to(device)
tokens['labels'] = tokens['input_ids'].clone()

def run_ablation_arm(indices):
    model = AutoModelForCausalLM.from_pretrained(MODEL_NAME, trust_remote_code=True, torch_dtype=torch.bfloat16).to(device)
    embed = model.transformer.wte if hasattr(model, "transformer") else model.model.embed_tokens

    # Snapshot initial state-space coordinates
    initial_matrix = embed.weight.detach().clone().float()
    cached_slots = {idx: embed.weight[idx].detach().clone() for idx in indices}

    # Freeze core block parameters to avoid VRAM overload
    for p in model.parameters(): p.requires_grad = False
    embed.weight.requires_grad = True
    optimizer = bnb.optim.AdamW8bit([embed.weight], lr=5e-4)

    model.train()
    for step in range(2):
        optimizer.zero_grad()
        loss = model(**tokens).loss
        loss.backward()
        optimizer.step()

        # Intercept and reset coordinates based on the active strategy group
        if indices:
            with torch.no_grad():
                for idx in indices:
                    embed.weight[idx].copy_(cached_slots[idx])

    # Quantify the global structural deformation delta
    model.eval()
    final_matrix = embed.weight.detach().clone().float()
    global_l2_displacement = torch.norm(final_matrix - initial_matrix).item()

    del model, optimizer
    gc.collect()
    torch.cuda.empty_cache()
    return global_l2_displacement

# Execute empirical evaluations sequentially
ablation_results = {}
for strategy_name, targets in ablation_strategies.items():
    print(f"[Ablation Lab] Launching baseline validation arm: {strategy_name}...")
    ablation_results[strategy_name] = run_ablation_arm(targets)

print("\n" + "=" * 80)
print("QUANTITATIVE ABLATION REPORT: GLOBAL EMBEDDING L2 DRIFT")
print("=" * 80)
for profile, metric in ablation_results.items():
    print(f"Strategy Layout: {profile:<30} | Global Embedding L2 Drift: {metric:.8f}")
print("=" * 80)
```

```python
import torch
import torch.nn as nn
import numpy as np
import gc
import bitsandbytes as bnb
from transformers import AutoModelForCausalLM, AutoTokenizer
from warnings import filterwarnings
filterwarnings('ignore')

# Force-inject local path for ast_lefm subsystem
import sys
sys.path.insert(0, '/content/ast_lefm')

try:
    from ast_lefm.sieve import primes_up_to
    from ast_lefm.h2e import h2e_sroi, h2e_decision
    print("✓ Native AST_LEFM library subsystem loaded successfully.")
except ImportError:
    raise ImportError("AST_LEFM library missing. Please run the git clone setup cells first.")

# =====================================================================
# 1. DETERMINISTIC REPRODUCIBILITY SEED CONFIGURATION
# =====================================================================
SEED = 123
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)
np.random.seed(SEED)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
MODEL_NAME = "openai/gpt-oss-20b"

print(f"Targeting active training on device: {device} (Deterministic Seed 123 Enabled)")

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, trust_remote_code=True)
if tokenizer.pad_token is None: tokenizer.pad_token = tokenizer.eos_token

# =====================================================================
# 2. GPT-OSS COMPATIBLE ACTIVE SPECTRAL GOVERNOR (CORRECTED FOR BF16)
# =====================================================================
class GPTOssActiveTrainingGovernor:
    def __init__(self, model, prime_limit=13):
        self.model = model
        self.primes = primes_up_to(prime_limit)
        self.LAMBDA_12 = 1.0 - np.prod([1.0 - (p ** -0.5) for p in self.primes])

        if hasattr(model, "transformer") and hasattr(model.transformer, "wte"):
            self.embed_layer = model.transformer.wte
        elif hasattr(model, "model") and hasattr(model.model, "embed_tokens"):
            self.embed_layer = model.model.embed_tokens
        else:
            raise AttributeError("Could not locate the gpt-oss token embedding matrix mapping.")

        print(f"✓ Governor linked to embedding matrix: {self.embed_layer.weight.shape}")

        self.cached_weights = {}
        for idx in self.primes:
            self.cached_weights[idx] = self.embed_layer.weight[idx].detach().clone().float()

    def enforce_topological_restoration(self):
        """Hard-locks anchor coordinates post-gradient step matching base tensor type."""
        with torch.no_grad():
            for idx in self.primes:
                target_dtype = self.embed_layer.weight.dtype
                self.embed_layer.weight[idx].copy_(self.cached_weights[idx].to(dtype=target_dtype))

# =====================================================================
# 3. METRIC TENSOR MATH ENGINE
# =====================================================================
def compute_local_metric_tensor(matrix, indices):
    """Calculates the local Riemannian metric tensor components g_ij."""
    vectors = matrix[indices].float()
    # g_ij = inner product space matrix <e_i, e_j>
    g_ij = torch.mm(vectors, vectors.t())
    return g_ij

# =====================================================================
# 4. RUNTIME VERIFICATION LOOP
# =====================================================================
# Load model layers in bfloat16 precision
model = AutoModelForCausalLM.from_pretrained(MODEL_NAME, trust_remote_code=True, torch_dtype=torch.bfloat16).to(device)
governor = GPTOssActiveTrainingGovernor(model, prime_limit=13)

# Isolate vocabulary embedding target weights for training
for param in model.parameters(): param.requires_grad = False
governor.embed_layer.weight.requires_grad = True
model.train()
optimizer = bnb.optim.AdamW8bit([governor.embed_layer.weight], lr=5e-4)

# Capture the baseline pristine geometry state profile
print("\n" + "=" * 80)
print("RIEMANNIAN SECTIONAL MANIFOLD METRIC TENSOR TRACKING")
print("=" * 80)
g_ij_initial = compute_local_metric_tensor(governor.embed_layer.weight, governor.primes)
det_initial = torch.det(g_ij_initial).item()
# FIX: Added .detach() before casting to cpu numpy to clear active gradient graph tracking
eigenvals_initial = torch.linalg.eigvalsh(g_ij_initial).detach().cpu().numpy()

print(f"Baseline Tensor Determinant det(g_ij)  : {det_initial:.6e}")
print(f"Baseline Manifold Isotropic Eigenvalues : {eigenvals_initial}\n")

# Process an active backpropagation step with optimization training data
sample_input = ["Arithmetic Spectral Theory preserves localized inner product spaces natively."]
tokens = tokenizer(sample_input, return_tensors="pt").to(device)
tokens['labels'] = tokens['input_ids'].clone()

optimizer.zero_grad()
loss = model(**tokens).loss
loss.backward()
optimizer.step()

# INTEGRATION LINK: Enforce native library restoration layer right post-gradient step
governor.enforce_topological_restoration()

# Recompute the post-gradient optimization state geometry profile
model.eval()
g_ij_trained = compute_local_metric_tensor(governor.embed_layer.weight, governor.primes)
det_trained = torch.det(g_ij_trained).item()
# FIX: Added .detach() here as well to cleanly isolate output arrays
eigenvals_trained = torch.linalg.eigvalsh(g_ij_trained).detach().cpu().numpy()

# Quantify structural deformation metrics
tensor_l2_drift = torch.norm(g_ij_trained - g_ij_initial).item()

print(f"Post-Train Tensor Determinant det(g_ij) : {det_trained:.6e}")
print(f"Post-Train Manifold Isotropic Eigenvals : {eigenvals_trained}")
print(f"-" * 80)
print(f"Absolute Inner Product Subspace Drift   : {tensor_l2_drift:.10f}")
print("=" * 80)

if tensor_l2_drift == 0.0:
    print("🎉 MATHEMATICAL PROOF CONFIRMED: Subspace Inner Products are perfectly Invariant!")
    print("The metric tensor g_ij remains unchanged, preserving local manifold topology.")
else:
    print(f"⚠️ Subspace Curvature Delta Tracked: Geometric variation is exactly {tensor_l2_drift:.10f}")
print("=" * 80)

# Clear device allocations safely
del model, optimizer
gc.collect()
torch.cuda.empty_cache()
```
